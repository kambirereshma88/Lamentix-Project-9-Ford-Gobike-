## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# 1. Import required libraries
import pandas as pd
import numpy as np

# 2. Load dataset
df = pd.read_csv('201801-fordgobike-tripdata.csv')

# 3. View first few records
print(df.head())

# 4. Check dataset info
df.info()

# 5. Convert 'start_time' and 'end_time' to datetime
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])

# 6. Create new features from datetime
df['start_date'] = df['start_time'].dt.date
df['start_hour'] = df['start_time'].dt.hour
df['day_of_week'] = df['start_time'].dt.day_name()
df['month'] = df['start_time'].dt.month

# 7. Create trip duration in minutes
df['duration_min'] = df['duration_sec'] / 60

# 8. Remove trips less than 1 minute or more than 4 hours (outlier filtering)
df = df[(df['duration_min'] >= 1) & (df['duration_min'] <= 240)]

# 9. Handle missing values
# Fill missing genders with 'Unknown'
df['member_gender'] = df['member_gender'].fillna('Unknown')

# Fill missing birth years with median (optional)
df['member_birth_year'] = df['member_birth_year'].fillna(df['member_birth_year'].median())

# 10. Remove duplicate rows if any
df.drop_duplicates(inplace=True)

# 11. Reset index
df.reset_index(drop=True, inplace=True)

# 12. Final check
df.info()
````
